```python
from pandas import read_csv, DataFrame
from os import path
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def plot_histplots(dataframe: DataFrame) -> None:
    """
    Plot Histograms for each column of the dataset:
    Useable code : sns.histplot(data=dataframe, x="Column name") 
    """
    for column in dataframe.columns:
        sns.histplot(data=dataframe, x=column)
        plt.show()
    return
        
def plot_histplot_grid(dataframe: DataFrame) -> None:
    """
    Plot a grid of histograms for the columns of the dataset
    Useable code :sns.histplot(dataframe, x="X column name", y="y column name")
    """
    grid = sns.PairGrid(dataframe)
    grid.map(sns.histplot) # scatterplot kdeplot
    return

def plot_kde_grid(dataframe: DataFrame) -> None:
    """
    Plot a grid of kde for the columns of the dataset
    Useable code :sns.kdeplot(dataframe, x="X column name", y="y column name")
    """
    grid = sns.PairGrid(dataframe)
    grid.map(sns.kdeplot) # scatterplot kdeplot
    return

def catplot_columns(dataframe: DataFrame) -> None:
    """
    Plot Catoplot for each numerical column of the dataset:
    Useable code : sns.catplot(data=dataframe, x="Column Name", kind="box") 
    """
    for column_name in dataframe.columns:
        if dataframe[column_name].dtype in ('int64', 'float64'):
            sns.catplot(data=dataframe, x=column_name, kind="box")
            plt.show()
    return None
```

```python
dataframe = read_csv(path.join("..", 'data/raw','carData.csv'))
dataframe.info()
```

```python
plot_histplots(dataframe)
```

```python
plot_histplot_grid(dataframe)
```

```python
plot_kde_grid(dataframe)
```

```python
catplot_columns(dataframe)
```
