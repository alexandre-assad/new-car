```python
from os import path

import matplotlib.pyplot as plt
from pandas import read_csv, DataFrame
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
def boxplot_price_by_categorical(dataframe: DataFrame) -> None:
    for column in dataframe.columns:
        if dataframe[column].dtype == 'object':
            sns.boxplot(data=dataframe, x="Selling_Price", y=column, width=.5)
            plt.show()

def violinplot_price_by_categorical(dataframe: DataFrame) -> None:
    for column in dataframe.columns:
        if dataframe[column].dtype == 'object':
            sns.violinplot(data=dataframe, x="Selling_Price", y=column, width=.5)
            plt.show()

def numerical_correlation_heatmap(dataframe: DataFrame) -> None:
    categorical_cols = [column for column in dataframe.columns if dataframe[column].dtype not in ('int64', 'float64') or column == 'Owner']
    dataframe_numerical = dataframe.drop(columns=categorical_cols)
    correlation_matrix = dataframe_numerical.corr() 
    sns.heatmap(data=correlation_matrix)

def global_correlation_pearson_heatmap(dataframe: DataFrame) -> None:
    encoded_dataframe = dataframe.copy()
    categorical_cols = [column for column in encoded_dataframe.columns if encoded_dataframe[column].dtype not in ('int64', 'float64') or column == 'Owner']
    label_encoder = LabelEncoder()

    for column in categorical_cols:
        encoded_dataframe[column] = label_encoder.fit_transform(dataframe[column])
    
    correlation_matrix = encoded_dataframe.corr() 
    sns.heatmap(data=correlation_matrix)
```

```python
dataframe = read_csv(path.join("..", 'data/raw','carData.csv'))
dataframe.info()
```

```python
numerical_correlation_heatmap(dataframe)
```

```python
global_correlation_pearson_heatmap(dataframe)
```

```python
boxplot_price_by_categorical(dataframe)
```

```python
violinplot_price_by_categorical(dataframe)
```
